# Food Security

There are over 200 definitions of food security however the most internationally accepted definition is; food security exists when all people at all-time have both physical and economic access to sufficient, safe and nutritious food that meets their dietary need and food preferences for an active and healthy life.Food security is a multidimensional concept that involves various factors that cut across social, political, environmental and economic issues. The four pillars of food security are:
1. Availability.
2. Access.
3. Utilization.
4. Stability.

The lack of a commonly accepted and comprehensive measure for food security on an international scale has been identified as one of the roadblocks on the way to the eradication of hunger and malnutrion. This research identifies a suit of indicators needed to cover the diferent dimensions of food security and uses a machine learning algorithm to generate a cluster of geographical zones of the 47 Counties in Kenya with similar food security scenarious.

## Dimensionality Reduction using PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Availability

```python
# Separating out the features
x = availability_df.loc[:, availability].values

# Standardizing the features
x = StandardScaler().fit_transform(x)

# PCA Analysis
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
availability_pca = pd.DataFrame(data = principalComponents, columns = ['availability'])


availability_pca.head()
```

### Accessibility

```python
# Separating out the features
x = accessibility_df.loc[:, accessibility].values

# Standardizing the features
x = StandardScaler().fit_transform(x)

# PCA Analysis
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
accessibility_pca = pd.DataFrame(data = principalComponents, columns = ['accessibility'])

accessibility_pca.head()
```

### Utilization

```python
# Separating out the features
x = utilization_df.loc[:, utilization].values

# Standardizing the features
x = StandardScaler().fit_transform(x)

# PCA Analysis
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
utilization_pca = pd.DataFrame(data = principalComponents, columns = ['utilization'])


utilization_pca.head()
```

### Stability

```python
# Separating out the features
x = stability_df.loc[:, stability].values

# Standardizing the features
x = StandardScaler().fit_transform(x)

# PCA Analysis
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
stability_pca = pd.DataFrame(data = principalComponents, columns = ['stability'])


stability_pca.head()
```

```python
# result = availability_pca.append([accessibility_pca, stability_pca])
food_security = pd.concat([fs_df['COUNTY'],fs_df[['geometry']],availability_pca,accessibility_pca, stability_pca,utilization_pca,], axis=1, join='inner')
food_security.head()
```

# CLustering

```python
from sklearn.mixture import GaussianMixture
```

```python
X=food_security[["availability","accessibility","stability","utilization"]]
EM = GaussianMixture(n_components=3)

EM.fit(X)

cluster = EM.predict(X)
# cluster
food_security['cluster']=cluster
```

```python
food_security.plot(column='cluster', cmap='OrRd', edgecolor='k')
```
